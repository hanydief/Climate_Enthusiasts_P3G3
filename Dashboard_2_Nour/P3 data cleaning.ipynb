```python
# !pip install Flask-SQLAlchemy
from flask import Flask, render_template
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)

# app.config['SQLALCHEMY_DATABASE_URI'] = 'postgresql://myuser:mypassword@localhost:5432/mydatabase'
app.config['SQLALCHEMY_DATABASE_URI'] = 'postgresql://postgres:MissPost$7@localhost:5432/P3_Climate_change'
db = SQLAlchemy(app)

# Define your model representing the climate_data table
class ClimateData(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    date = db.Column(db.DateTime)
    location = db.Column(db.String(255))
    country = db.Column(db.String(255))
    temperature = db.Column(db.Float)
    co2_emissions = db.Column(db.Float)
    sea_level_rise = db.Column(db.Float)
    precipitation = db.Column(db.Float)
    humidity = db.Column(db.Float)
    wind_speed = db.Column(db.Float)

# Create a route to display the climate data
@app.route('/')
def display_climate_data():
    print("Running the querry")
    data = ClimateData.query.all()
    return data

# render_template('climate_data.html', data=data)

if __name__ == '__main__':
    app.run()
```
